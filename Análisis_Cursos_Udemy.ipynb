```python
#Buscamos el coeficiente de correlacion entre 'avg_rating' y 'avg_rating_recent'
correlacion_co = stats.pearsonr(df_udemy2['avg_rating'], df_udemy2['avg_rating_recent'])[0]
np.around(correlacion_co, 2)
```

El coeficiente de correlación es 0,99. Esto significa que podemos predecir 'avg_rating_recent' a partir de 'avg_rating'. Por lo tanto, eliminaremos la columna 'avg_rating_recent'.

```python
#Eliminamos la columna avg_rating_recent
df_udemy2 = df_udemy2.drop(columns=['avg_rating_recent'])
```



1.   Todos los cursos gratuitos tienen menos de 50.000 suscriptores con 3 excepciones.

```python
# Creación de diagramas de dispersión entre todas las demás variables numéricas con cursos gratuitos/de pago codificados por colores
(fig, ((p1, p2, p3), (p4, p5, p6), (p7, p8, p9))) = plt.subplots(3, 3, figsize=(15,14))

p1.scatter(df_udemy2['num_subscribers'], df_udemy2['avg_rating'], c=df_udemy2['is_paid'].astype('category').cat.codes, cmap='Set1')
p1.set_title('Suscriptores y Ratings')

p2.scatter(df_udemy2['num_subscribers'], df_udemy2['num_published_lectures'], c=df_udemy2['is_paid'].astype('category').cat.codes, cmap='Set1')
p2.set_title('Suscriptores y Lecturas')

p3.scatter(df_udemy2['num_subscribers'], df_udemy2['price'], c=df_udemy2['is_paid'].astype('category').cat.codes, cmap='Set1')
p3.set_title('Suscriptores y Precio')

p4.scatter(df_udemy2['avg_rating'], df_udemy2['num_reviews'], c=df_udemy2['is_paid'].astype('category').cat.codes, cmap='Set1')
p4.set_title('Ratings y Reviews')

p5.scatter(df_udemy2['avg_rating'], df_udemy2['num_published_lectures'], c=df_udemy2['is_paid'].astype('category').cat.codes, cmap='Set1')
p5.set_title('Ratings y Lecturas')

p6.scatter(df_udemy2['avg_rating'], df_udemy2['price'], c=df_udemy2['is_paid'].astype('category').cat.codes, cmap='Set1')
p6.set_title('Ratings y Precio')

p7.scatter(df_udemy2['num_reviews'], df_udemy2['num_published_lectures'], c=df_udemy2['is_paid'].astype('category').cat.codes, cmap='Set1')
p7.set_title('Reviews y Lecturas')

p8.scatter(df_udemy2['num_reviews'], df_udemy2['price'], c=df_udemy2['is_paid'].astype('category').cat.codes, cmap='Set1')
p8.set_title('Reviews y Precio')

p9.scatter(df_udemy2['num_published_lectures'], df_udemy2['price'], c=df_udemy2['is_paid'].astype('category').cat.codes, cmap='Set1')
p9.set_title('Lecturas y Precio')

plt.show()
```



1.   Todos los diagramas de dispersión que contienen la variable precio son fáciles de interpretar, ya que todos los cursos gratuitos se muestran en Precio = 0.
